# Linear Support Vector Machine 

```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Train dataset
X = pd.read_csv('Xsvm.csv', header=None)
y = pd.read_csv('ysvm.csv', header=None)
x_train = np.asarray(X)
y_train = np.asarray(y).flatten()
x_train.shape, y_train.shape
```

```python
# test data
x_test = np.array([[2,0.5],
                   [0.8,0.7],
                   [1.58,1.33],
                   [0.008,0.001]])
```

##### Observing the training dataset. 
We are given 2-D vectors with labels belonging to two classes y = 1 and y = -1. Plotting these 2-D vectors in a 2-D plane. 

```python
dummy = np.argsort(y_train) # sorted labels for further use
x1 = np.array([x_train[i] for i in dummy[-np.sum(y_train==1):]]).T
x_1 = np.array([x_train[i] for i in dummy[:np.sum(y_train==-1)]]).T
plt.plot(x1[0], x1[1], '*', x_1[0], x_1[1],'*')
plt.show()
```
