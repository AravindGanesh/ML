# Maximum Likelihood Estimator
1. Binomial
2. Poisson
3. Exponential
4. Gaussian
5. Laplacian

```python
import numpy as np
import matplotlib.pyplot as plt
N = int(1e5)  # number of data points
```

## Perform MLE on data sampled from same distribution
Take random samples from a distribution (using some parameters) and find optimal parameters for maximum likelihood. Compare this data with random samples generated with optimal parameters.

#### Binomial

```python
p = 0.3
X = np.random.binomial(1,p, N)
p_ml = np.mean(X)
Y = np.random.binomial(1,p_ml, N)

#plot histograms
fig = plt.figure(figsize=(50,10))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100)
fig.add_subplot(1,2,2) # generated samples
plt.hist(Y, bins=100)
plt.show()
```

#### Poisson 

```python
Lambda = 10
X = np.random.poisson(Lambda, N)
Lambda_ml = np.mean(X)
Y = np.random.poisson(Lambda_ml, N)

#plot histograms
fig = plt.figure(figsize=(50,10))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100)
fig.add_subplot(1,2,2) # generated samples
plt.hist(Y, bins=100)
plt.show()
```

#### Exponential 

```python
a = 3
X = np.random.exponential(a, N)
a_ml = 1/np.mean(X)
Y = np.random.exponential(a_ml, N)

#plot histograms
fig = plt.figure(figsize=(50,10))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100)
fig.add_subplot(1,2,2) # generated samples
plt.hist(Y, bins=100)
plt.show()
```

#### Gaussian 

```python
mu, sigma = 0 , 1
X = np.random.normal(mu,sigma, N)
mu_ml, sigma_ml = np.mean(X), np.std(X)
Y = np.random.normal(mu_ml,sigma_ml, N)

#plot histograms
fig = plt.figure(figsize=(50,10))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100)
fig.add_subplot(1,2,2) # generated samples
plt.hist(Y, bins=100)
plt.show()
```

#### Laplace

```python
Mu, b = 0 , 3
X = np.random.laplace(Mu,b, N)
Mu_ml = np.median(X)
b_ml = np.mean(np.abs(X-Mu))
Y = np.random.laplace(Mu_ml,b_ml, N)

#plot histograms
fig = plt.figure(figsize=(50,10))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100)
fig.add_subplot(1,2,2) # generated samples
plt.hist(Y, bins=100)
plt.show()
```

## Perform MLE on random sampled data

##### Uniform data

```python
X = np.random.uniform(0,1,N)
```

```python
fig = plt.figure(figsize=(20,10))
fig.add_subplot(2,3,1)
plt.hist(X, bins=100, label='Original - Uniform')
# Binomial estimator
p_ml = np.mean(X)
Y1 = np.random.binomial(1,p_ml, N)
fig.add_subplot(2,3,2) 
plt.hist(Y1, bins=100, label='Binomial')
# Poisson estimator
Lambda_ml = np.mean(X)
Y2 = np.random.poisson(Lambda_ml, N)
fig.add_subplot(2,3,3) 
plt.hist(Y2, bins=100, label='Poisson')
# Exponential Estimator
a_ml = 1/np.mean(X)
Y3 = np.random.exponential(a_ml, N)
fig.add_subplot(2,3,4) 
plt.hist(Y3, bins=100, label='Exponential')
# Gaussian Estimator
mu_ml, sigma_ml = np.mean(X), np.std(X)
Y4 = np.random.normal(mu_ml,sigma_ml, N)
fig.add_subplot(2,3,5) 
plt.hist(Y4, bins=100, label='Normal')
# Laplacian Estimator
Mu_ml = np.median(X)
b_ml = np.mean(np.abs(X-Mu))
Y5 = np.random.laplace(Mu_ml,b_ml, N)
fig.add_subplot(2,3,6) 
plt.hist(Y5, bins=100, label='Laplacian')
plt.show()
```

##### Gaussian Data
