# Law of Large Numbers 

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Uniform 

```python
N = int(1e3)
x_uniform = np.random.uniform(low=0, high=1, size=N)
X = np.cumsum(x_uniform)/ np.arange(1,N+1)
print(np.mean(x_uniform))
fig = plt.figure(figsize=(50,20))
plt.plot(X)
plt.show()
```

### Binomial

```python
N = int(1e3)

x_binomial = np.random.binomial(n=1, p=0.3, size=N)
X = np.cumsum(x_binomial)/np.arange(1,N+1)
print(np.mean(x_binomial))
fig = plt.figure(figsize=(50,20))
plt.plot(X)
plt.show()
```

### Laplacian 

```python
N = int(1e3)

x_laplacian = np.random.laplace(loc=0, scale=1, size=N)
X = np.cumsum(x_laplacian)/np.arange(1,N+1)
print(np.mean(x_laplacian))
fig = plt.figure(figsize=(50,20))
plt.plot(X)
plt.show()
```

### Poisson 

```python
N = int(1e3)

x_poisson = np.random.poisson(lam=1, size=N)
X = np.cumsum(x_poisson)/np.arange(1,N+1)
print(np.mean(x_poisson))
fig = plt.figure(figsize=(50,20))
plt.plot(X)
plt.show()
```

# Central Limit Theorem 

```python
def normal_pdf(x, mu, var):
    px = ((2*np.pi*var)**(-0.5)) * np.exp(-((x-mu)**2)/(2*var))
    return px
```

```python
x = np.linspace(start=-0.1, stop=0.1, num=10)
normal_pdf(x, 0, 1)
```

### Uniform 

```python
N = int(1e3)

x_uniform = np.random.uniform(low=-1, high=1, size=N)
X = np.cumsum(x_uniform)/np.arange(1,N+1)
mean, var = np.mean(x_uniform), np.var(x_uniform)
y = np.linspace(start=np.min(X), stop=np.max(X), num=N)
Y = normal_pdf(y, mean, var)
#plot histograms
fig = plt.figure(figsize=(50,20))
fig.add_subplot(1,2,1) # given data points
plt.hist(X, bins=100, density=True)
fig.add_subplot(1,2,2) # generated samples
plt.plot(Y)
plt.show()
```
