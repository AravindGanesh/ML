```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate training samples
N = 10 # Number of training samples
# Generate equispaced floats in the interval [0, 2Ï€]
x = np.linspace(0, 2*np.pi, N)
# Generate noise
mean = 0
std = 0.05
# Generate some numbers from the sine function
y = np.sin(x)
# Add noise
y += np.random.normal(mean, std, N)
```

```python
# Generate test inputs
N_test = 100
x_test = np.linspace(0,2*np.pi, N_test)
test_mean = 0
test_std = 0.05
x_test += np.random.normal(test_mean, test_std, N_test)
```

### _l2_ Regularized Linear Regression using Polynomial basis function

```python
def get_regularized_weights(X, _y, _lambda):
    w = np.dot(np.dot(np.linalg.inv(np.dot(np.transpose(X),X) + _lambda*np.identity(len(np.transpose(X)))), np.transpose(X)), _y)
    return w
```

```python
def polynomial(X, _M):
    _phi_x = X
    for i in range(2, _M+1):
        _phi_x = np.c_[_phi_x, X**i]
    return _phi_x
```

```python
# without regularization
M = 4
l = 0.0
y1 = y - np.mean(y) # make mean zero so that we do not need to use bias
x_train = polynomial(x, M)
# print(x_train.shape)
w1 = get_regularized_weights(x_train,y1, l)
# print(w1.shape)
y_predict = np.dot(polynomial(x_test, M), w1) + np.mean(y)
# plotting
plt.plot(x,y,'bo')
plt.plot(x_test, y_predict, 'ro')
plt.show() 
```

```python
# without regularization 
M = 13
l = 0.0
y1 = y - np.mean(y) # make mean zero so that we do not need to use bias
x_train = polynomial(x, M)
# print(x_train.shape)
w1 = get_regularized_weights(x_train,y1, l)
# print(w1.shape)
y_predict = np.dot(polynomial(x_test, M), w1) + np.mean(y)
# plotting
plt.plot(x,y,'bo')
plt.plot(x_test, y_predict, 'ro')
plt.show() 
```

Curve tries to _over-fit_ as order of the polynomial increases as shown in the plot. When there is less data and more number a parameters, model tends to over-fit i.e. fails to generalize the input - label relation.

```python
# above cell, with regularization
M = 13
l = 0.6
y1 = y - np.mean(y) # make mean zero so that we do not need to use bias
x_train = polynomial(x, M)
# print(x_train.shape)
w1 = get_regularized_weights(x_train,y1, l)
# print(w1.shape)
y_predict = np.dot(polynomial(x_test, M), w1) + np.mean(y)
# plotting
plt.plot(x,y,'bo')
plt.plot(x_test, y_predict, 'ro')
plt.show() 
```

l2 regularization - constraint on norm of weights vector is applied. This resulted in a better fit.
